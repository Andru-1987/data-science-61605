# Desafío Genérico

```python
import pandas as pd
```

```python
customers= pd.read_csv('customers.csv',sep=',')
customers.head()
```

```python
orders= pd.read_csv('orders.csv',sep=',')
orders.head()
```

```python
employees= pd.read_csv('employees.csv',sep=',')
employees.head()
```

1) Obtener un dataframe con el OrderID, LastName y FirstName con base en las tablas orders y employees.

```python
pd.merge(orders, employees, how='inner', on='EmployeeID').sort_values(by='OrderID', ascending=False)[['OrderID','LastName','FirstName']]
```

2) Obtener un dataframe con el CustomerName y el OrderID utilizando las tablas Customers y Orders

```python
pd.merge(orders, customers, how='inner', on='CustomerID').sort_values(by='OrderID')[['OrderID','CustomerName']]
```
